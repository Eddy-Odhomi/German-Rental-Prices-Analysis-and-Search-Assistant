```python
import pandas as pd
import os

# Use the correct file path
directory = r"C:\Users\odhomi\Desktop\Data Analysis\Bootcamp WBS\Final Project\CSV Files"  # Change to your directory

# List all CSV files in the directory
csv_files = [file for file in os.listdir(directory) if file.endswith('.csv')]
print(csv_files)  # Optional: Print the list of files to verify

# Initialize an empty list to hold all DataFrames
dataframes = []

# Loop through each CSV file and read its data into a DataFrame
for file in csv_files:
    file_path = os.path.join(directory, file)  # Create full file path
    data = pd.read_csv(file_path)  # Read the CSV file
    dataframes.append(data)  # Append DataFrame to the list

# Concatenate all DataFrames into a single DataFrame
merged_data = pd.concat(dataframes, ignore_index=True)

# Optional: Print the shape of the merged DataFrame
print(merged_data.shape)

# Save the merged DataFrame to a new CSV file
merged_data.to_csv('merged_data.csv', index=False)

# Read the merged data for verification
merged_data = pd.read_csv('merged_data.csv')
merged_data  # Display the first few rows
```

```python
merged_data_df = merged_data.copy()
```

```python
schema = "rental_data"
host = "127.0.0.1"
user = "root"
password = "password"
port = 3306

connection_string = f'mysql+pymysql://{user}:{password}@{host}:{port}/{schema}'
```

```python
from sqlalchemy import create_engine

# Create an engine for connecting to MySQL
engine = create_engine(connection_string)

# Test the connection
with engine.connect() as connection:
    print("Connection successful")
```

```python
# Transfer the DataFrame to MySQL (will create a new table if it doesn't exist)
merged_data_df.to_sql('german_cities', con=engine, if_exists='replace', index=False)
```
