```python
import pandas as pd
from googlemaps import Client
import time

def get_postal_code(address, city, api_key):
    gmaps = Client(key=api_key)
    full_address = f"{address}, {city}, Germany"
    try:
        time.sleep(0.1)  # Add a small delay to avoid hitting rate limits
        result = gmaps.geocode(full_address)
        if result:
            for component in result[0]['address_components']:
                if 'postal_code' in component['types']:
                    return component['long_name']
    except Exception as e:
        print(f"Error geocoding {full_address}: {e}")
    return None

# Replace 'YOUR_API_KEY' with your actual Google Maps API key
api_key = 'Your API Key'

# Assuming your DataFrame is named hamburg_data
frankfurt_data['postal_code'] = frankfurt_data.apply(
    lambda row: get_postal_code(row['Street'], row['City'], api_key)
    if row['Street'] != 'N/A' else None,
    axis=1
)
```

## Stuttgart with both Temporary and Permanent columns

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

def clean_date(date_string):
    """Clean the date string."""
    cleaned = re.sub(r'\s+', ' ', date_string).strip()
    if ' - ' in cleaned:
        start, end = cleaned.split(' - ')
        return f"{start.strip()} - {end.strip()}"
    return cleaned

def get_wg_gesucht_data(city, pages=4):
    all_data = []
    for page in range(1, pages + 1):
        url = f"https://www.wg-gesucht.de/1-zimmer-wohnungen-in-Stuttgart.124.1.1.0.html"
        headers = {
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
        }
        response = requests.get(url, headers=headers)
        
        # Check response status
        if response.status_code != 200:
            print(f"Failed to retrieve page {page}. Status code: {response.status_code}")
            continue
        
        soup = BeautifulSoup(response.content, 'html.parser')
        
        listings = soup.find_all('div', class_='wgg_card offer_list_item')
        if not listings:
            print(f"No listings found on page {page}.")
            continue
        
        for listing in listings:
            title_element = listing.find('h3', class_='truncate_title')
            title_tag = title_element.find('a') if title_element else None
            title = title_tag.get_text(strip=True) if title_tag else 'N/A'
            listing_url = f"https://www.wg-gesucht.de{title_tag['href']}" if title_tag else 'N/A'
            
            listing_id = listing_url.split('.')[-2] if listing_url != 'N/A' else 'N/A'
            
            wg_info_element = listing.find('span', text=True)
            wg_type, location = 'N/A', 'N/A'
            if wg_info_element:
                wg_info_text = wg_info_element.get_text(separator="|", strip=True)
                wg_info_parts = wg_info_text.split('|')
                if wg_info_parts:
                    wg_type = wg_info_parts[0].strip()
                    if len(wg_info_parts) > 1:
                        location = ' | '.join(wg_info_parts[1:]).strip()
            
            rent_price_element = listing.find('b')
            rent_price = rent_price_element.get_text(strip=True) if rent_price_element and '€' in rent_price_element.get_text() else 'N/A'
            
            area_element = listing.find_all('b')
            m2 = 'N/A'
            for area in area_element:
                if 'm²' in area.get_text():
                    m2 = area.get_text(strip=True)
                    break
            
            date_element = listing.find('div', class_='col-xs-5 text-center')
            listed_date = date_element.get_text(strip=True) if date_element else 'N/A'
            
            owner_element = listing.find('span', class_='ml5')
            owner = owner_element.get_text(strip=True) if owner_element else 'N/A'
            
            all_data.append({
                'id': listing_id,
                'title': title,
                'wg_type': wg_type,
                'location': location,
                'rent_price': rent_price,
                'area': m2,
                'listed_date': listed_date,
                'owner': owner,
            })
    
    stuttgart_data = pd.DataFrame(all_data)
    
    # Check if 'listed_date' exists before proceeding
    if 'listed_date' not in stuttgart_data.columns:
        print("The 'listed_date' column is missing from the data.")
        return stuttgart_data
    
    # Clean the listed_date column
    stuttgart_data['available_date'] = stuttgart_data['listed_date'].apply(clean_date)

    # Clean the location column and split into parts
    stuttgart_data['location_clean'] = stuttgart_data['location'].str.replace(r'\n', '', regex=True).str.strip()
    location_split = stuttgart_data['location_clean'].str.split('|', n=1, expand=True)

    stuttgart_data['City'] = location_split[0].str.extract(r'^(.*?)\s')[0].str.strip()
    stuttgart_data['Neighbourhood'] = location_split[0].str.extract(r'\s(.*)$')[0].str.strip()
    stuttgart_data['Street'] = location_split[1].str.strip().replace(r'\n', '', regex=True).fillna('N/A')
    
    stuttgart_data['rental_type'] = stuttgart_data['available_date'].apply(lambda x: 'Temporary' if ' - ' in x else 'Long-term')
    
    # Drop the original listed_date column if no longer needed
    stuttgart_data = stuttgart_data.drop('listed_date', axis=1)
    stuttgart_data = stuttgart_data.drop(['location', 'location_clean'], axis=1)
    
    return stuttgart_data

# Example usage:
city = 'Stuttgart'
stuttgart_data = get_wg_gesucht_data(city)
#print(stuttgart_data.head(10))
stuttgart_data.head(5)
```

### Geo coding usage

```python
import pandas as pd
from googlemaps import Client
import time

def get_postal_code(address, city, api_key):
    gmaps = Client(key=api_key)
    full_address = f"{address}, {city}, Germany"
    try:
        time.sleep(0.1)  # Add a small delay to avoid hitting rate limits
        result = gmaps.geocode(full_address)
        if result:
            for component in result[0]['address_components']:
                if 'postal_code' in component['types']:
                    return component['long_name']
    except Exception as e:
        print(f"Error geocoding {full_address}: {e}")
    return None

# Replace 'YOUR_API_KEY' with your actual Google Maps API key
api_key = 'API KEY'

# Assuming your DataFrame is named hamburg_data
stuttgart_data['postal_code'] = stuttgart_data.apply(
    lambda row: get_postal_code(row['Street'], row['City'], api_key)
    if row['Street'] != 'N/A' else None,
    axis=1
)

```
